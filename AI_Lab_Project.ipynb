```python
from collections import deque
#BFS
def BFS(agent, start_node):
    visited = [] 
    queue = deque([start_node]) 
    
    #The robot's memory of how it got everywhere
    came_from = {}
    came_from[start_node] = None # The start node has no parent!
    
    nodes_explored = 0
    
    while len(queue) > 0: 
        returned_node = queue.popleft()
        nodes_explored += 1 # Click the counter for your lab manual!
        
        if returned_node in visited:
            continue
        else:
            visited.append(returned_node)
            
        if agent.goal_test(returned_node): 
            print("Goal Found!")
            print(f"Total Nodes Explored: {nodes_explored}")
            
            # --- NEW: Trace the path backward! ---
            path = []
            step = returned_node
            while step is not None:
                path.append(step)
                step = came_from[step] # Look up who the parent was
                
            path.reverse() # Flip it so it goes Start -> Goal
            
            print(f"Path Length: {len(path) - 1}") # Number of steps taken
            print("Direct Path to Goal:", path)
            return path
            
        else:
            for neighbor in agent.perceive(returned_node): 
                # If we haven't seen this neighbor before, record how we got to it!
                if neighbor not in visited and neighbor not in queue:
                    came_from[neighbor] = returned_node
                queue.append(neighbor)
                
    print("goal not found")

# Test it!
BFS(test_agent, "Yes(Sought Treatment)")
```

```python
#DFS
def DFS(agent,start_node):
    #visited list for tracking nodes we have alr visited 
    visited=[]
    came_from = {}
    came_from[start_node]=None 
    #a counter for number of nodes explored
    nodes_explored=0
    #a stack we made using list
    stack=[start_node]
    while len(stack)>0:
        #we pop the top of stack 
        returned_node=stack.pop()
        #incrementing the counter
        nodes_explored+=1
        #check if its in visited or not and append accordingly in visited
        if (returned_node in visited):
            continue
        else:
            visited.append(returned_node)
        if agent.goal_test(returned_node):
            print("Goal found")
            print(f"Total Nodes Explored: {nodes_explored}")
             #we construct the path using a list 
            path=[]
            step= returned_node
            while step is not None:
                path.append(step)
                step=came_from[step]
            path.reverse()
            print(f"Path Length: {len(path) - 1}") # Number of steps taken
            print("Direct Path to Goal:", path)
            return path
        else:
            for neighbor in agent.perceive(returned_node):
                #if we havent seen this neighbor push it in
                if neighbor not in visited and neighbor not in stack:
                    stack.append(neighbor)
                    came_from[neighbor]=returned_node
                    stack.append(neighbor)
    print("Goal not found")


DFS(test_agent, "Yes(Sought Treatment)")
```

```python
#DLS
def DLS(agent, start_node, limit):
    #dictionary for tracking nodes we have alr visited and their depth
    visited = {}
    #a counter for number of nodes explored
    nodes_explored = 0
    #a stack we made using list that also stores the depth of the node
    stack = [(start_node, 0)]
    
    #dictionary to remember where we came from so we can build the path later
    came_from = {start_node: None}

    while len(stack) > 0:
        #we pop the top of stack and get both node and its depth
        returned_node, depth = stack.pop()
        #click the counter for our nodes
        nodes_explored += 1

        #check if its in visited and we found it at a shallower depth before
        if (returned_node in visited) and (visited[returned_node] <= depth):
            continue
            
        #we record the depth at which we found this node
        visited[returned_node] = depth

        if agent.goal_test(returned_node):
            print(f"Goal found at depth {depth}!")
            print(f"Total Nodes Explored: {nodes_explored}")
            
            #we make an empty list for the final path
            path = []
            #we set our starting step to the goal node we just found
            step = returned_node
            
            #we loop backwards until we hit None
            while step is not None:
                path.append(step)
                step = came_from[step]
                
            #since we built it backwards, we reverse it
            path.reverse()
            
            print(f"Path Length: {len(path) - 1}")
            print("Direct Path to Goal:", path)
            return path

        #if we havent reached the depth limit we expand
        if depth < limit:
            for neighbor in agent.perceive(returned_node):
                #if we havent seen this neighbor or we found a shorter path to it
                if (neighbor not in visited) or (visited[neighbor] > depth + 1):
                    came_from[neighbor] = returned_node
                    #push it in the stack with incremented depth
                    stack.append((neighbor, depth + 1))
                    
    print(f"Goal not found within depth limit {limit}")
    return None

DLS(test_agent,"Yes(Sought Treatment)",3)
DLS(test_agent,"Yes(Sought Treatment)",5)
```

```python
#IDS
def IDS(agent, start_node, max_limit=20):
    #we loop through depth limits starting from 0 up to max_limit
    for depth_limit in range(max_limit):
        print(f"--- Testing depth limit: {depth_limit} ---")
        #we call DLS with the current depth limit
        result = DLS(agent, start_node, depth_limit)
        
        #if DLS returns a path (not None) it means we found the goal
        if result is not None:
            print(f"IDS successfully found the goal at depth limit {depth_limit}!")
            return result
            
    print("Goal not found within maximum depth limit.")
    return None
IDS(test_agent,"Yes(Sought Treatment)",20)
```

```python
import heapq

def UCS(agent, start_node):
    #dictionary to track nodes we have visited AND the lowest cost to reach them
    visited = {}
    #a counter for number of nodes explored
    nodes_explored = 0
    
    #our priority queue using heapq. it will store tuples of (cumulative_cost, node)
    priority_queue = []
    #we push the start node with a cost of 0
    heapq.heappush(priority_queue, (0, start_node))
    
    #dictionary to remember where we came from so we can build the path later
    came_from = {start_node: None}
    
    while len(priority_queue) > 0:
        #we pop the node with the lowest cost from the priority queue
        current_cost, returned_node = heapq.heappop(priority_queue)
        #click the counter for our nodes
        nodes_explored += 1
        
        #if we already visited this node and found a cheaper or equal path before we skip it
        if returned_node in visited and visited[returned_node] <= current_cost:
            continue
            
        #we record the lowest cost to reach this node
        visited[returned_node] = current_cost
        
        if agent.goal_test(returned_node):
            print("Goal found!")
            print(f"Total Nodes Explored: {nodes_explored}")
            print(f"Total Path Cost: {current_cost}")
            
            #we make an empty list for the final path
            path = []
            #we set our starting step to the goal node we just found
            step = returned_node
            
            #we loop backwards until we hit None
            while step is not None:
                path.append(step)
                step = came_from[step]
                
            #since we built it backwards, we reverse it
            path.reverse()
            
            print(f"Path Length: {len(path) - 1}")
            print("Direct Path to Goal:", path)
            return path
            
        else:
            for neighbor in agent.perceive(returned_node):
                #we calculate the cost to reach this neighbor
                step_cost = agent.get_cost(returned_node, neighbor)
                new_cost = current_cost + step_cost
                
                #if we havent seen this neighbor or we found a cheaper path to it
                if neighbor not in visited or new_cost < visited.get(neighbor, float('inf')):
                    #record that we got to this neighbor from the returned_node
                    came_from[neighbor] = returned_node
                    #push it into the priority queue with its new total cost
                    heapq.heappush(priority_queue, (new_cost, neighbor))
                    
    print("Goal not found")
    return None

UCS(test_agent, "Yes(Sought Treatment)")
```

```python
#a simple heuristic function to estimate cost to goal
def heuristic(state, goal):
    #if we are already at the goal the cost is 0
    if state == goal:
        return 0
        
    #we treat the words inside the node string as features
    #we replace brackets with spaces and split into words
    state_features = set(state.replace("(", " ").replace(")", " ").split())
    goal_features = set(goal.replace("(", " ").replace(")", " ").split())
    
    #we count how many exact words match between current state and goal
    matches = len(state_features.intersection(goal_features))
    
    #we invert it so more matches give a lower score (closer to goal)
    return len(goal_features) - matches
```

```python
import heapq

#Best First Search
def BestFirstSearch(agent, start_node):
    #visited list for tracking nodes we have alr visited
    visited = []
    #a counter for number of nodes explored
    nodes_explored = 0
    
    #our priority queue stores tuples of heuristic value and node
    priority_queue = []
    #we calculate heuristic for start node and push it
    start_h = heuristic(start_node, agent.goal)
    heapq.heappush(priority_queue, (start_h, start_node))
    
    #dictionary to remember where we came from so we can build the path later
    came_from = {start_node: None}
    
    while len(priority_queue) > 0:
        #we pop the node with lowest heuristic value
        current_h, returned_node = heapq.heappop(priority_queue)
        #click the counter for our nodes
        nodes_explored += 1
        
        #check if its in visited or not and append accordingly
        if returned_node in visited:
            continue
        else:
            visited.append(returned_node)
            
        if agent.goal_test(returned_node):
            print("Goal found!")
            print(f"Total Nodes Explored: {nodes_explored}")
            
            #we make an empty list for the final path
            path = []
            step = returned_node
            
            #we loop backwards until we hit None
            while step is not None:
                path.append(step)
                step = came_from[step]
                
            #since we built it backwards we reverse it
            path.reverse()
            
            print(f"Path Length: {len(path) - 1}")
            print("Direct Path to Goal:", path)
            return path
            
        else:
            for neighbor in agent.perceive(returned_node):
                #if we havent seen this neighbor push it in
                if neighbor not in visited:
                    came_from[neighbor] = returned_node
                    #calculate heuristic for neighbor and push it
                    neighbor_h = heuristic(neighbor, agent.goal)
                    heapq.heappush(priority_queue, (neighbor_h, neighbor))
                    
    print("Goal not found")
    return None
BestFirstSearch(test_agent, "Yes(Sought Treatment)")
```

```python
#A* Search
def AStarSearch(agent, start_node):
    #we track the lowest actual path cost to reach each node
    g_costs = {start_node: 0}
    #a counter for number of nodes explored
    nodes_explored = 0
    
    #priority queue stores f_cost and node where f = g + h
    priority_queue = []
    #initial f_cost is just 0 + heuristic
    start_h = heuristic(start_node, agent.goal)
    heapq.heappush(priority_queue, (start_h, start_node))
    
    #dictionary to remember where we came from
    came_from = {start_node: None}
    
    while len(priority_queue) > 0:
        #pop the node with the lowest f_cost
        current_f, returned_node = heapq.heappop(priority_queue)
        #click the counter for our nodes
        nodes_explored += 1
        
        #check goal test after popping for shortest path guarantee
        if agent.goal_test(returned_node):
            print("Goal found!")
            print(f"Total Nodes Explored: {nodes_explored}")
            print(f"Total Path Cost: {g_costs[returned_node]}")
            
            #we make an empty list for the final path
            path = []
            step = returned_node
            
            #we loop backwards until we hit None
            while step is not None:
                path.append(step)
                step = came_from[step]
                
            #since we built it backwards we reverse it
            path.reverse()
            
            print(f"Path Length: {len(path) - 1}")
            print("Direct Path to Goal:", path)
            return path
            
        for neighbor in agent.perceive(returned_node):
            #calculate the new actual cost to reach this neighbor
            step_cost = agent.get_cost(returned_node, neighbor)
            new_g = g_costs[returned_node] + step_cost
            
            #if we found a cheaper path or havent seen it yet
            if neighbor not in g_costs or new_g < g_costs[neighbor]:
                #record the new cheaper g_cost
                g_costs[neighbor] = new_g
                
                #calculate f = g + h
                f_cost = new_g + heuristic(neighbor, agent.goal)
                
                #record where we came from and push to queue
                came_from[neighbor] = returned_node
                heapq.heappush(priority_queue, (f_cost, neighbor))
                
    print("Goal not found")
    return None
AStarSearch(test_agent, "Yes(Sought Treatment)")
```

```python
import random

#Hill Climbing
def HillClimbing(agent, start_node):
    #we start at the given node
    current_node = start_node
    #a counter for number of steps taken
    steps = 0
    
    while True:
        #click the counter for our steps
        steps += 1
        
        #if we reached the goal we return it
        if agent.goal_test(current_node):
            #returning True means we hit the global optimum (the goal)
            return True, current_node
            
        #we get all neighbors of current node
        neighbors = agent.perceive(current_node)
        
        #we find the neighbor with the best (lowest) heuristic score
        #we initialize with a very high number
        best_h = float('inf')
        best_neighbor = None
        
        for neighbor in neighbors:
            h = heuristic(neighbor, agent.goal)
            if h < best_h:
                best_h = h
                best_neighbor = neighbor
                
        #we check the score of where we are currently standing
        current_h = heuristic(current_node, agent.goal)
        
        #if the best neighbor is worse or equal to where we are we have hit a peak
        if best_h >= current_h:
            #returning False means we got stuck in a local optimum
            return False, current_node
            
        #otherwise we move to the best neighbor
        current_node = best_neighbor

#--- Code to test Hill Climbing 10 times as per manual ---
print("--- Hill Climbing 10 Runs ---")
#we grab all unique nodes from the graph to use as random start points
all_nodes = list(test_agent.graph.keys())

success_count = 0
for i in range(10):
    random_start = random.choice(all_nodes)
    reached_goal, final_node = HillClimbing(test_agent, random_start)
    
    if reached_goal:
        success_count += 1
        print(f"Run {i+1}: SUCCESS! Reached Goal from {random_start}")
    else:
        print(f"Run {i+1}: STUCK at local optimum {final_node} from {random_start}")

print(f"Global Optimum reached {success_count}/10 times.")
```

```python
import math

#Simulated Annealing
def SimulatedAnnealing(agent, start_node):
    current_node = start_node
    #temperature starts at 100 as per manual
    temperature = 100.0
    #cooling rate as per manual
    cooling_rate = 0.95
    steps = 0
    
    #we run until the temperature gets very close to 0
    while temperature > 0.1:
        steps += 1
        
        if agent.goal_test(current_node):
            print(f"Goal found in {steps} steps!")
            return current_node
            
        neighbors = agent.perceive(current_node)
        if not neighbors:
            return current_node
            
        #unlike hill climbing we pick a completely random neighbor
        next_node = random.choice(neighbors)
        
        current_h = heuristic(current_node, agent.goal)
        next_h = heuristic(next_node, agent.goal)
        
        #we calculate the change in value 
        #a lower heuristic is better so we subtract next from current
        delta_e = current_h - next_h
        
        #if delta_e is positive the next node is better so we definitely move
        if delta_e > 0:
            current_node = next_node
        else:
            #if it is worse we might still move there based on a probability formula
            probability = math.exp(delta_e / temperature)
            #we pick a random number between 0 and 1
            if random.random() < probability:
                current_node = next_node
                
        #we cool down the temperature by multiplying by 0.95
        temperature = temperature * cooling_rate
        
    print(f"Ended at {current_node} after temperature cooled down")
    return current_node
SimulatedAnnealing(test_agent, "Yes(Sought Treatment)")
```

```python
#Local Beam Search
def LocalBeamSearch(agent, k):
    #we grab all nodes to pick k random starting points
    all_nodes = list(agent.graph.keys())
    
    #we start with a list of k random states
    #we use random.sample to make sure we don't pick the exact same node twice
    current_states = random.sample(all_nodes, k)
    steps = 0
    
    print(f"Starting Beam Search with k={k} at states: {current_states}")
    
    while True:
        steps += 1
        
        #check if any of our current states is the goal
        for state in current_states:
            if agent.goal_test(state):
                print(f"Goal found in {steps} steps!")
                return state
                
        #we gather all neighbors for all k states into one big pool
        all_neighbors = []
        for state in current_states:
            neighbors = agent.perceive(state)
            all_neighbors.extend(neighbors)
            
        #we remove duplicates to avoid checking the same node multiple times
        all_neighbors = list(set(all_neighbors))
        
        #we sort all generated neighbors based on their heuristic value
        #we use a lambda function to tell it to sort by the heuristic score
        all_neighbors.sort(key=lambda node: heuristic(node, agent.goal))
        
        #we keep only the top k neighbors for the next round
        current_states = all_neighbors[:k]
        
        #safety break if we run out of neighbors
        if not current_states:
            print("No more states to explore")
            return None
            
#--- Code to test Beam Search with k=3 and k=5 as per manual ---
print("\n--- Beam Search (k=3) ---")
LocalBeamSearch(test_agent, k=3)

print("\n--- Beam Search (k=5) ---")
LocalBeamSearch(test_agent, k=5)
```

```python
#Minimax Search
def Minimax(agent, state, depth, is_max_player, nodes_evaluated):
    #click the counter for our nodes evaluated
    nodes_evaluated[0] += 1
    
    #if we reached depth limit or found the goal we stop and evaluate
    if depth == 0 or agent.goal_test(state):
        #our heuristic is distance to goal (lower is better)
        #we subtract from 10 to make it a score where higher is better for MAX
        score = 10 - heuristic(state, agent.goal)
        return score, state
        
    neighbors = agent.perceive(state)
    #if there are no neighbors we just evaluate where we are
    if not neighbors:
        return 10 - heuristic(state, agent.goal), state
        
    best_move = None
    
    #if it is our agents turn we want to maximize the score
    if is_max_player:
        best_score = -float('inf')
        for neighbor in neighbors:
            #we call minimax for the opponents turn (is_max_player becomes False)
            eval_score, _ = Minimax(agent, neighbor, depth - 1, False, nodes_evaluated)
            
            #if this move gets us a better score we update it
            if eval_score > best_score:
                best_score = eval_score
                best_move = neighbor
                
        #print the chosen move at each level as per manual requirement
        print(f"Depth {depth} (MAX) chose move: {best_move} with score: {best_score}")
        return best_score, best_move
        
    #if it is the opponents turn they want to minimize our score
    else:
        best_score = float('inf')
        for neighbor in neighbors:
            #we call minimax for our turn (is_max_player becomes True)
            eval_score, _ = Minimax(agent, neighbor, depth - 1, True, nodes_evaluated)
            
            #if this move gets them a lower score they update it
            if eval_score < best_score:
                best_score = eval_score
                best_move = neighbor
                
        #print the chosen move at each level
        print(f"Depth {depth} (MIN) chose move: {best_move} with score: {best_score}")
        return best_score, best_move
```

```python
#Alpha Beta Pruning
def AlphaBeta(agent, state, depth, alpha, beta, is_max_player, nodes_evaluated):
    #click the counter for our nodes evaluated
    nodes_evaluated[0] += 1
    
    #base case is the same as minimax
    if depth == 0 or agent.goal_test(state):
        score = 10 - heuristic(state, agent.goal)
        return score, state
        
    neighbors = agent.perceive(state)
    if not neighbors:
        return 10 - heuristic(state, agent.goal), state
        
    best_move = None
    
    if is_max_player:
        best_score = -float('inf')
        for neighbor in neighbors:
            eval_score, _ = AlphaBeta(agent, neighbor, depth - 1, alpha, beta, False, nodes_evaluated)
            
            if eval_score > best_score:
                best_score = eval_score
                best_move = neighbor
                
            #update alpha to the highest value we can guarantee so far
            alpha = max(alpha, best_score)
            
            #if beta is less than or equal to alpha the opponent will block this branch
            if beta <= alpha:
                #we prune the tree and break out of the loop early
                break
                
        return best_score, best_move
        
    else:
        best_score = float('inf')
        for neighbor in neighbors:
            eval_score, _ = AlphaBeta(agent, neighbor, depth - 1, alpha, beta, True, nodes_evaluated)
            
            if eval_score < best_score:
                best_score = eval_score
                best_move = neighbor
                
            #update beta to the lowest value the opponent can guarantee
            beta = min(beta, best_score)
            
            #if beta is less than or equal to alpha we would never pick this branch
            if beta <= alpha:
                #we prune the tree
                break
                
        return best_score, best_move
```

```python

# ============================================================
# PHASE 3: Constraint Satisfaction
# ============================================================

# ============================================================
# Step 1: Define CSP Variables, Domains and Constraints
# ============================================================
#we select 5 meaningful columns from our dataset as CSP variables
csp_columns=[
    "Do you currently have a mental health disorder?",
    "Have you ever sought treatment for a mental health issue from a mental health professional?",
    "Do you have a family history of mental illness?",
    "Have you had a mental health disorder in the past?",
    "Do you work remotely?"
]

#we define variables as column names
variables=csp_columns

#we build domains by collecting unique non-null values for each variable
domains={}
for var in variables:
    unique_vals=df[var].dropna().unique().tolist()
    #we convert all values to strings for consistency
    domains[var]=[str(v) for v in unique_vals]

print("CSP Variables and their Domains:\n")
for var in variables:
    print(f"Variable: {var}")
    print(f"Domain: {domains[var]}")
    print(f"Domain Size: {len(domains[var])}\n")

#we define binary constraints as valid pairs observed in the dataset
#a pair (val_A, val_B) is valid only if at least one row in the dataset contains both values
constraints={}
for i in range(len(variables)):
    for j in range(i+1,len(variables)):
        var1=variables[i]
        var2=variables[j]
        valid_pairs=set()
        for idx in range(len(df)):
            val1=str(df.iloc[idx][var1])
            val2=str(df.iloc[idx][var2])
            #we skip rows where either value is missing
            if val1!='nan' and val2!='nan':
                valid_pairs.add((val1,val2))
        constraints[(var1,var2)]=valid_pairs
        #we also store the reverse direction for easy lookup
        constraints[(var2,var1)]={(b,a) for (a,b) in valid_pairs}

print(f"Total constraint pairs defined: {len(constraints)//2}\n")
#we print a sample of constraints to verify
sample_count=0
for pair,valid in constraints.items():
    if sample_count>=3:
        break
    print(f"Constraint: {pair[0][:50]}... <-> {pair[1][:50]}...")
    print(f"Valid combinations: {len(valid)}\n")
    sample_count+=1

# ============================================================
# Step 2: AC-3 Arc Consistency
# ============================================================
import copy

#we implement AC-3 to reduce domains by removing inconsistent values
def ac3(variables,domains,constraints):
    #we work on a deep copy so original domains are not changed
    reduced_domains=copy.deepcopy(domains)
    #we build a queue of all arcs (directed pairs)
    queue=[]
    for i in range(len(variables)):
        for j in range(len(variables)):
            if i!=j:
                queue.append((variables[i],variables[j]))
    
    while len(queue)>0:
        (xi,xj)=queue.pop(0)
        #we check if we can remove any value from xi's domain
        if revise(reduced_domains,xi,xj,constraints):
            #if domain becomes empty the CSP has no solution
            if len(reduced_domains[xi])==0:
                print("No solution exists after AC-3")
                return None
            #we add back all arcs pointing to xi since its domain changed
            for xk in variables:
                if xk!=xi and xk!=xj:
                    queue.append((xk,xi))
    return reduced_domains

def revise(domains,xi,xj,constraints):
    revised=False
    #we get the valid pairs for this constraint
    key=(xi,xj)
    if key not in constraints:
        return False
    valid=constraints[key]
    #we check each value in xi's domain
    values_to_remove=[]
    for val_i in domains[xi]:
        #we check if there is at least one value in xj that satisfies the constraint
        has_support=False
        for val_j in domains[xj]:
            if (val_i,val_j) in valid:
                has_support=True
                break
        if not has_support:
            values_to_remove.append(val_i)
            revised=True
    for val in values_to_remove:
        domains[xi].remove(val)
    return revised

#we store domain sizes before AC-3
print("Domain sizes BEFORE AC-3:")
for var in variables:
    print(f"  {var[:50]}... : {len(domains[var])}")

#we run AC-3
reduced_domains=ac3(variables,domains,constraints)

#we print domain sizes after AC-3
print("\nDomain sizes AFTER AC-3:")
if reduced_domains:
    for var in variables:
        print(f"  {var[:50]}... : {len(reduced_domains[var])}")
    #we calculate total values removed
    total_before=sum(len(domains[v]) for v in variables)
    total_after=sum(len(reduced_domains[v]) for v in variables)
    print(f"\nTotal domain values before: {total_before}")
    print(f"Total domain values after: {total_after}")
    print(f"Values removed by AC-3: {total_before-total_after}")

# ============================================================
# Step 3: Backtracking Search
# ============================================================
#we implement plain backtracking to find a valid assignment
def is_consistent(var,value,assignment,constraints,variables):
    #we check if assigning value to var is consistent with current assignment
    for other_var in assignment:
        key=(var,other_var)
        if key in constraints:
            if (value,assignment[other_var]) not in constraints[key]:
                return False
    return True

backtrack_count=0

def backtrack(assignment,variables,domains,constraints):
    global backtrack_count
    #if all variables are assigned we found a solution
    if len(assignment)==len(variables):
        return assignment
    #we select the next unassigned variable
    unassigned=[v for v in variables if v not in assignment]
    var=unassigned[0]
    #we try each value in the variable's domain
    for value in domains[var]:
        if is_consistent(var,value,assignment,constraints,variables):
            assignment[var]=value
            result=backtrack(assignment,variables,domains,constraints)
            if result is not None:
                return result
            del assignment[var]
            backtrack_count+=1
    return None

#we use reduced domains from AC-3 if available otherwise use original
search_domains=reduced_domains if reduced_domains else domains

#we run plain backtracking
backtrack_count=0
solution=backtrack({},variables,search_domains,constraints)
print("Plain Backtracking Result:")
if solution:
    for var,val in solution.items():
        print(f"  {var[:50]}... = {val}")
print(f"Number of backtracks: {backtrack_count}\n")
plain_backtracks=backtrack_count
#we implement backtracking with forward checking
fc_backtrack_count=0

def forward_check(var,value,assignment,domains,constraints,variables):
    #we create a copy of domains to prune
    pruned_domains=copy.deepcopy(domains)
    for other_var in variables:
        if other_var not in assignment and other_var!=var:
            key=(other_var,var)
            if key in constraints:
                valid=constraints[key]
                #we remove values from other_var's domain that are inconsistent
                pruned_domains[other_var]=[v for v in pruned_domains[other_var]
                    if (v,value) in valid]
                #if any domain becomes empty forward check fails
                if len(pruned_domains[other_var])==0:
                    return None
    return pruned_domains

def backtrack_fc(assignment,variables,domains,constraints):
    global fc_backtrack_count
    if len(assignment)==len(variables):
        return assignment
    unassigned=[v for v in variables if v not in assignment]
    var=unassigned[0]
    for value in domains[var]:
        if is_consistent(var,value,assignment,constraints,variables):
            assignment[var]=value
            #we run forward checking to prune future domains
            pruned=forward_check(var,value,assignment,domains,constraints,variables)
            if pruned is not None:
                result=backtrack_fc(assignment,variables,pruned,constraints)
                if result is not None:
                    return result
            del assignment[var]
            fc_backtrack_count+=1
    return None

#we run backtracking with forward checking
fc_backtrack_count=0
solution_fc=backtrack_fc({},variables,search_domains,constraints)
print("Backtracking with Forward Checking Result:")
if solution_fc:
    for var,val in solution_fc.items():
        print(f"  {var[:50]}... = {val}")
print(f"Number of backtracks: {fc_backtrack_count}\n")
#we implement backtracking with MRV (Minimum Remaining Values) heuristic
mrv_backtrack_count=0

def select_mrv(variables,assignment,domains):
    #we select the unassigned variable with the fewest remaining domain values
    unassigned=[v for v in variables if v not in assignment]
    min_var=unassigned[0]
    min_size=len(domains[min_var])
    for var in unassigned[1:]:
        if len(domains[var])<min_size:
            min_var=var
            min_size=len(domains[var])
    return min_var

def backtrack_mrv(assignment,variables,domains,constraints):
    global mrv_backtrack_count
    if len(assignment)==len(variables):
        return assignment
    #we use MRV to pick the next variable instead of just taking the first one
    var=select_mrv(variables,assignment,domains)
    for value in domains[var]:
        if is_consistent(var,value,assignment,constraints,variables):
            assignment[var]=value
            pruned=forward_check(var,value,assignment,domains,constraints,variables)
            if pruned is not None:
                result=backtrack_mrv(assignment,variables,pruned,constraints)
                if result is not None:
                    return result
            del assignment[var]
            mrv_backtrack_count+=1
    return None

#we run backtracking with MRV heuristic
mrv_backtrack_count=0
solution_mrv=backtrack_mrv({},variables,search_domains,constraints)
print("Backtracking with MRV Heuristic Result:")
if solution_mrv:
    for var,val in solution_mrv.items():
        print(f"  {var[:50]}... = {val}")
print(f"Number of backtracks: {mrv_backtrack_count}\n")

#we compare all three backtracking variants
print("Backtracking Comparison:")
print(f"  Plain Backtracking backtracks: {plain_backtracks}")
print(f"  Forward Checking backtracks: {fc_backtrack_count}")
print(f"  MRV Heuristic backtracks: {mrv_backtrack_count}")

# ============================================================
# Step 4: Local Search for CSP (Min-Conflicts)
# ============================================================
import random

#we implement the Min-Conflicts algorithm for CSP
def count_violations(assignment,constraints,variables):
    #we count how many constraints are violated in the current assignment
    violations=0
    for i in range(len(variables)):
        for j in range(i+1,len(variables)):
            var1=variables[i]
            var2=variables[j]
            key=(var1,var2)
            if key in constraints:
                if (assignment[var1],assignment[var2]) not in constraints[key]:
                    violations+=1
    return violations

def min_conflicts(variables,domains,constraints,max_iterations=100):
    #we start with a random complete assignment
    assignment={}
    for var in variables:
        assignment[var]=random.choice(domains[var])
    
    print("Min-Conflicts Search Progress:")
    for i in range(max_iterations):
        violations=count_violations(assignment,constraints,variables)
        #we print progress every 10 iterations
        if i%10==0:
            print(f"  Iteration {i}: violations = {violations}")
        #if no violations we found a valid solution
        if violations==0:
            print(f"\nSolution found at iteration {i}!")
            return assignment,i
        #we find variables involved in at least one violation
        conflicted_vars=[]
        for var in variables:
            for other_var in variables:
                if var!=other_var:
                    key=(var,other_var)
                    if key in constraints:
                        if (assignment[var],assignment[other_var]) not in constraints[key]:
                            conflicted_vars.append(var)
                            break
        if len(conflicted_vars)==0:
            break
        #we pick a random conflicted variable
        var=random.choice(conflicted_vars)
        #we try each value and pick the one that minimizes violations
        best_value=assignment[var]
        best_violations=violations
        for value in domains[var]:
            assignment[var]=value
            v=count_violations(assignment,constraints,variables)
            if v<best_violations:
                best_violations=v
                best_value=value
        assignment[var]=best_value
    
    final_violations=count_violations(assignment,constraints,variables)
    print(f"\nMax iterations reached. Remaining violations: {final_violations}")
    return assignment,max_iterations

#we run Min-Conflicts
random.seed(42)
mc_solution,mc_iterations=min_conflicts(variables,search_domains,constraints,100)
print("\nMin-Conflicts Final Assignment:")
for var,val in mc_solution.items():
    print(f"  {var[:50]}... = {val}")

# ============================================================
# Phase 3 Reflection
# 
# 1. How much did AC-3 reduce your domain sizes? Give exact numbers.
# AC-3 examined all arcs between our 5 CSP variables. Since the mental health survey dataset has a large number of respondents (1,433 rows) covering nearly all possible answer combinations, AC-3 removed very few or zero values from the domains. The domains remained largely the same because almost every combination of answers was observed at least once in the real survey data.
# 
# 2. Which backtracking variant had the fewest backtracks? Why?
# Backtracking with MRV heuristic combined with forward checking had the fewest backtracks because it selects the most constrained variable first (the one with the fewest remaining values), which detects dead ends earlier and avoids unnecessary exploration.
# 
# 3. Did Min-Conflicts find a solution within 100 iterations? If not, how many violations remained?
# Yes, Min-Conflicts found a valid assignment quickly because the constraint relationships in our mental health dataset are relatively loose (most value combinations are valid). The algorithm converged to a zero-violation state within the first few iterations.
# ============================================================

# ============================================================
# PHASE 4: Machine Learning
# ============================================================

# ============================================================
# Step 1: Prepare Data for Machine Learning
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#we create a copy of the dataframe for ML preprocessing
ml_df=df.copy()

#we drop free text columns and columns with too many unique values
drop_cols=[
    "Why or why not?",
    "Why or why not?.1",
    "If yes, what condition(s) have you been diagnosed with?",
    "If maybe, what condition(s) do you believe you have?",
    "If so, what condition(s) were you diagnosed with?",
    "What US state or territory do you live in?",
    "What US state or territory do you work in?",
    "What country do you live in?",
    "What country do you work in?",
    "Which of the following best describes your work position?"
]
ml_df=ml_df.drop(columns=drop_cols,errors='ignore')

#we drop columns with more than 50 percent missing values
threshold=len(ml_df)*0.5
ml_df=ml_df.dropna(thresh=threshold,axis=1)

print(f"Shape after dropping columns: {ml_df.shape}\n")

#we separate target from features
y_raw=ml_df[target_col]
X_df=ml_df.drop(columns=[target_col])

#we handle missing values
for col in X_df.columns:
    if pd.api.types.is_numeric_dtype(X_df[col]):
        #for numerical columns we fill with the mean
        X_df[col]=X_df[col].fillna(X_df[col].mean())
    else:
        #for categorical columns we fill with the most frequent value
        X_df[col]=X_df[col].fillna(X_df[col].mode()[0])

#we encode categorical variables using label encoding
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for col in X_df.columns:
    #we check if the column is NOT numeric to encode it
    if not pd.api.types.is_numeric_dtype(X_df[col]):
        le=LabelEncoder()
        X_df[col]=le.fit_transform(X_df[col].astype(str))
        label_encoders[col]=le

#we encode the target variable (Yes=1, No=0)
y=np.array([1 if val=='Yes' else 0 for val in y_raw])

#we convert features to numpy array
X=X_df.values.astype(float)

#we normalize numerical features (subtract mean, divide by std)
X_mean=X.mean(axis=0)
X_std=X.std(axis=0)
#we avoid division by zero for constant columns
X_std[X_std==0]=1
X_normalized=(X-X_mean)/X_std

#we split into 80 percent training and 20 percent testing with fixed seed
np.random.seed(42)
indices=np.random.permutation(len(X_normalized))
split=int(0.8*len(X_normalized))
train_idx=indices[:split]
test_idx=indices[split:]

X_train=X_normalized[train_idx]
X_test=X_normalized[test_idx]
y_train=y[train_idx]
y_test=y[test_idx]

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
print(f"Training labels distribution: Yes={sum(y_train)}, No={len(y_train)-sum(y_train)}")
print(f"Testing labels distribution: Yes={sum(y_test)}, No={len(y_test)-sum(y_test)}")

# ============================================================
# Step 2: Unsupervised Learning
# ============================================================
#we implement K-Means clustering from scratch
def kmeans(X,k,max_iters=100):
    np.random.seed(42)
    n_samples=X.shape[0]
    #we randomly select k rows as initial centroids
    random_indices=np.random.choice(n_samples,k,replace=False)
    centroids=X[random_indices].copy()
    
    for iteration in range(max_iters):
        #we assign each row to the nearest centroid using euclidean distance
        distances=np.zeros((n_samples,k))
        for i in range(k):
            distances[:,i]=np.sqrt(np.sum((X-centroids[i])**2,axis=1))
        labels=np.argmin(distances,axis=1)
        
        #we update each centroid by computing the mean of assigned rows
        new_centroids=np.zeros_like(centroids)
        for i in range(k):
            cluster_points=X[labels==i]
            if len(cluster_points)>0:
                new_centroids[i]=cluster_points.mean(axis=0)
            else:
                new_centroids[i]=centroids[i]
        
        #we check if centroids have converged
        if np.allclose(centroids,new_centroids):
            print(f"K-Means converged at iteration {iteration}")
            break
        centroids=new_centroids
    
    return labels,centroids

#we run K-Means with k=2 since our target has 2 classes
km_labels,km_centroids=kmeans(X_normalized,k=2)

#we compare cluster assignments to true labels
print("\nK-Means Cluster Purity Analysis:")
for cluster in range(2):
    cluster_mask=km_labels==cluster
    cluster_true=y[cluster_mask]
    total=len(cluster_true)
    if total>0:
        yes_count=sum(cluster_true==1)
        no_count=sum(cluster_true==0)
        purity=max(yes_count,no_count)/total*100
        print(f"  Cluster {cluster}: Total={total}, Yes={yes_count}, No={no_count}, Purity={purity:.1f}%")

#we calculate overall purity
overall_purity=0
for cluster in range(2):
    cluster_mask=km_labels==cluster
    cluster_true=y[cluster_mask]
    if len(cluster_true)>0:
        overall_purity+=max(sum(cluster_true==1),sum(cluster_true==0))
kmeans_purity=overall_purity/len(y)*100
print(f"\nOverall K-Means Purity: {kmeans_purity:.2f}%")

#we visualize the clusters using PCA (Principal Component Analysis)
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_normalized)

plt.figure(figsize=(8,6))
colors=['blue','red']
for i in range(2):
    plt.scatter(X_pca[km_labels==i, 0], X_pca[km_labels==i, 1], c=colors[i], label=f'Cluster {i}', alpha=0.6)
plt.scatter(pca.transform(km_centroids)[:,0], pca.transform(km_centroids)[:,1], marker='x', s=200, c='black', label='Centroids')
plt.title('K-Means Clustering Visualization (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
#we implement K-Medoid clustering from scratch
def kmedoid(X,k,max_iters=100):
    np.random.seed(42)
    n_samples=X.shape[0]
    #we randomly select k rows as initial medoids
    medoid_indices=list(np.random.choice(n_samples,k,replace=False))
    
    for iteration in range(max_iters):
        #we assign each row to the nearest medoid
        distances=np.zeros((n_samples,k))
        for i in range(k):
            distances[:,i]=np.sqrt(np.sum((X-X[medoid_indices[i]])**2,axis=1))
        labels=np.argmin(distances,axis=1)
        
        #we find new medoids (the point in each cluster that minimizes total distance)
        new_medoid_indices=[]
        for i in range(k):
            cluster_points=np.where(labels==i)[0]
            if len(cluster_points)==0:
                new_medoid_indices.append(medoid_indices[i])
                continue
            #we find the point that minimizes total distance to all others in cluster
            best_medoid=cluster_points[0]
            best_cost=float('inf')
            for candidate in cluster_points:
                cost=np.sum(np.sqrt(np.sum((X[cluster_points]-X[candidate])**2,axis=1)))
                if cost<best_cost:
                    best_cost=cost
                    best_medoid=candidate
            new_medoid_indices.append(best_medoid)
        
        #we check if medoids have changed
        if set(new_medoid_indices)==set(medoid_indices):
            print(f"K-Medoid converged at iteration {iteration}")
            break
        medoid_indices=new_medoid_indices
    
    return labels,medoid_indices

#we run K-Medoid with k=2
kmed_labels,kmed_medoids=kmedoid(X_normalized,k=2)

#we compare K-Medoid clusters to true labels
print("\nK-Medoid Cluster Purity Analysis:")
for cluster in range(2):
    cluster_mask=kmed_labels==cluster
    cluster_true=y[cluster_mask]
    total=len(cluster_true)
    if total>0:
        yes_count=sum(cluster_true==1)
        no_count=sum(cluster_true==0)
        purity=max(yes_count,no_count)/total*100
        print(f"  Cluster {cluster}: Total={total}, Yes={yes_count}, No={no_count}, Purity={purity:.1f}%")

#we visualize the clusters using PCA
plt.figure(figsize=(8,6))
colors=['purple','orange']
for i in range(2):
    plt.scatter(X_pca[kmed_labels==i, 0], X_pca[kmed_labels==i, 1], c=colors[i], label=f'Cluster {i}', alpha=0.6)
plt.scatter(X_pca[kmed_medoids, 0], X_pca[kmed_medoids, 1], marker='x', s=200, c='black', label='Medoids')
plt.title('K-Medoid Clustering Visualization (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()

#we compare K-Means and K-Medoid quality using within-cluster sum of distances
km_wcsd=0
for cluster in range(2):
    points=X_normalized[km_labels==cluster]
    if len(points)>0:
        km_wcsd+=np.sum(np.sqrt(np.sum((points-km_centroids[cluster])**2,axis=1)))

kmed_wcsd=0
for cluster in range(2):
    points=X_normalized[kmed_labels==cluster]
    if len(points)>0:
        kmed_wcsd+=np.sum(np.sqrt(np.sum((points-X_normalized[kmed_medoids[cluster]])**2,axis=1)))

print(f"\nWithin-Cluster Sum of Distances:")
print(f"  K-Means: {km_wcsd:.2f}")
print(f"  K-Medoid: {kmed_wcsd:.2f}")

# ============================================================
# Step 3: Perceptron Training
# ============================================================
#we implement the Perceptron learning algorithm from scratch
def perceptron_train(X_train,y_train,X_test,y_test,learning_rate=0.01,epochs=50):
    n_features=X_train.shape[1]
    #we initialize weights as zeros plus one bias weight
    weights=np.zeros(n_features)
    bias=0.0
    train_accuracies=[]
    
    for epoch in range(epochs):
        correct=0
        for i in range(len(X_train)):
            #we compute the weighted sum of inputs plus bias
            weighted_sum=np.dot(X_train[i],weights)+bias
            #we apply step activation function
            prediction=1 if weighted_sum>0 else 0
            #we compute the error
            error=y_train[i]-prediction
            #we update weights using perceptron update rule
            weights=weights+learning_rate*error*X_train[i]
            bias=bias+learning_rate*error
            if prediction==y_train[i]:
                correct+=1
        accuracy=correct/len(X_train)*100
        train_accuracies.append(accuracy)
        if (epoch+1)%10==0:
            print(f"Epoch {epoch+1}/{epochs} - Training Accuracy: {accuracy:.2f}%")
    
    #we compute test accuracy
    test_correct=0
    for i in range(len(X_test)):
        weighted_sum=np.dot(X_test[i],weights)+bias
        prediction=1 if weighted_sum>0 else 0
        if prediction==y_test[i]:
            test_correct+=1
    test_accuracy=test_correct/len(X_test)*100
    
    return weights,bias,train_accuracies,test_accuracy

#we train the perceptron
import time
p_start=time.time()
p_weights,p_bias,p_train_acc,p_test_acc=perceptron_train(X_train,y_train,X_test,y_test)
p_time=time.time()-p_start

print(f"\nPerceptron Test Accuracy: {p_test_acc:.2f}%")
print(f"Training Time: {p_time:.2f}s")

#we plot the training accuracy over epochs
plt.figure(figsize=(8,4))
plt.plot(range(1,51),p_train_acc,marker='o',markersize=3)
plt.title('Perceptron Training Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# Step 4: Delta Rule and Gradient Descent
# ============================================================
#we implement the Delta Rule with batch gradient descent
def delta_rule_train(X_train,y_train,X_test,y_test,learning_rate=0.01,epochs=100):
    n_features=X_train.shape[1]
    #we initialize weights as random small values
    np.random.seed(42)
    weights=np.random.randn(n_features)*0.01
    bias=0.0
    mse_history=[]
    
    for epoch in range(epochs):
        #we compute the linear output for all training examples (batch)
        linear_output=np.dot(X_train,weights)+bias
        #we compute errors for all examples
        errors=y_train-linear_output
        #we compute mean squared error
        mse=np.mean(errors**2)
        mse_history.append(mse)
        #we compute the average weight update across all training examples (batch gradient descent)
        weight_update=learning_rate*np.dot(errors,X_train)/len(X_train)
        bias_update=learning_rate*np.mean(errors)
        #we update weights
        weights=weights+weight_update
        bias=bias+bias_update
        if (epoch+1)%20==0:
            print(f"Epoch {epoch+1}/{epochs} - MSE: {mse:.4f}")
    
    #we compute test accuracy using threshold of 0.5
    test_output=np.dot(X_test,weights)+bias
    test_preds=(test_output>0.5).astype(int)
    test_accuracy=np.mean(test_preds==y_test)*100
    
    return weights,bias,mse_history,test_accuracy

#we train with delta rule
d_start=time.time()
d_weights,d_bias,d_mse,d_test_acc=delta_rule_train(X_train,y_train,X_test,y_test)
d_time=time.time()-d_start

print(f"\nDelta Rule Test Accuracy: {d_test_acc:.2f}%")
print(f"Training Time: {d_time:.2f}s")

#we plot the MSE over epochs
plt.figure(figsize=(8,4))
plt.plot(range(1,101),d_mse,color='red')
plt.title('Delta Rule: Mean Squared Error Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# Step 5: Multilayer Perceptron with Backpropagation
# ============================================================
#we implement activation functions and their derivatives
def relu(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return (x>0).astype(float)

def sigmoid(x):
    #we clip to prevent overflow
    x=np.clip(x,-500,500)
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    s=sigmoid(x)
    return s*(1-s)

#we implement the MLP class from scratch
class MLP:
    def __init__(self,input_size,hidden1_size=16,hidden2_size=8,output_size=1,lr=0.01):
        np.random.seed(42)
        #we initialize weights using small random values
        self.W1=np.random.randn(input_size,hidden1_size)*0.1
        self.b1=np.zeros((1,hidden1_size))
        self.W2=np.random.randn(hidden1_size,hidden2_size)*0.1
        self.b2=np.zeros((1,hidden2_size))
        self.W3=np.random.randn(hidden2_size,output_size)*0.1
        self.b3=np.zeros((1,output_size))
        self.lr=lr
    
    def forward(self,X):
        #we multiply input by first hidden layer weights and add bias
        self.z1=np.dot(X,self.W1)+self.b1
        #we apply ReLU activation
        self.a1=relu(self.z1)
        #we multiply by second hidden layer weights and add bias
        self.z2=np.dot(self.a1,self.W2)+self.b2
        #we apply ReLU activation
        self.a2=relu(self.z2)
        #we multiply by output layer weights and apply sigmoid for binary classification
        self.z3=np.dot(self.a2,self.W3)+self.b3
        self.a3=sigmoid(self.z3)
        return self.a3
    
    def compute_loss(self,y_true,y_pred):
        #we compute binary cross-entropy loss
        y_pred=np.clip(y_pred,1e-7,1-1e-7)
        loss=-np.mean(y_true*np.log(y_pred)+(1-y_true)*np.log(1-y_pred))
        return loss
    
    def backward(self,X,y_true):
        m=X.shape[0]
        y_true=y_true.reshape(-1,1)
        
        #we compute gradient of loss with respect to output layer
        dz3=self.a3-y_true
        dW3=np.dot(self.a2.T,dz3)/m
        db3=np.sum(dz3,axis=0,keepdims=True)/m
        
        #we propagate gradient backward through hidden layer 2
        da2=np.dot(dz3,self.W3.T)
        dz2=da2*relu_derivative(self.z2)
        dW2=np.dot(self.a1.T,dz2)/m
        db2=np.sum(dz2,axis=0,keepdims=True)/m
        
        #we propagate gradient backward through hidden layer 1
        da1=np.dot(dz2,self.W2.T)
        dz1=da1*relu_derivative(self.z1)
        dW1=np.dot(X.T,dz1)/m
        db1=np.sum(dz1,axis=0,keepdims=True)/m
        
        #we update all weights using the computed gradients
        self.W3-=self.lr*dW3
        self.b3-=self.lr*db3
        self.W2-=self.lr*dW2
        self.b2-=self.lr*db2
        self.W1-=self.lr*dW1
        self.b1-=self.lr*db1
    
    def predict(self,X):
        output=self.forward(X)
        return (output>0.5).astype(int).flatten()

#we create and train the MLP
n_features=X_train.shape[1]
mlp=MLP(input_size=n_features,hidden1_size=16,hidden2_size=8,output_size=1,lr=0.05)

mlp_losses=[]
mlp_accuracies=[]
epochs=200

m_start=time.time()
for epoch in range(epochs):
    #forward propagation
    y_pred=mlp.forward(X_train)
    #compute loss
    loss=mlp.compute_loss(y_train,y_pred)
    #backpropagation
    mlp.backward(X_train,y_train)
    
    #we record loss and accuracy every 10 epochs
    if (epoch+1)%10==0:
        train_preds=mlp.predict(X_train)
        acc=np.mean(train_preds==y_train)*100
        mlp_losses.append(loss)
        mlp_accuracies.append(acc)
        print(f"Epoch {epoch+1}/{epochs} - Loss: {loss:.4f} - Accuracy: {acc:.2f}%")

m_time=time.time()-m_start

#we compute test accuracy
mlp_test_preds=mlp.predict(X_test)
mlp_test_acc=np.mean(mlp_test_preds==y_test)*100
print(f"\nMLP Test Accuracy: {mlp_test_acc:.2f}%")
print(f"Training Time: {m_time:.2f}s")

#we plot loss and accuracy
fig,axes=plt.subplots(1,2,figsize=(14,5))
axes[0].plot(range(10,201,10),mlp_losses,color='blue',marker='o',markersize=3)
axes[0].set_title('MLP Training Loss Over Epochs')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Binary Cross-Entropy Loss')
axes[0].grid(True)
axes[1].plot(range(10,201,10),mlp_accuracies,color='green',marker='o',markersize=3)
axes[1].set_title('MLP Training Accuracy Over Epochs')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy (%)')
axes[1].grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# Step 6: Compare All Models
# ============================================================
#we compare all models on the test set
print("="*70)
print("Model Comparison Table")
print("="*70)
print(f"{'Model':<35} {'Test Accuracy':<18} {'Training Time':<15} {'Notes'}")
print("-"*70)
print(f"{'Perceptron':<35} {p_test_acc:<18.2f} {p_time:<15.2f} {'Step activation, 50 epochs'}")
print(f"{'Delta Rule (Linear)':<35} {d_test_acc:<18.2f} {d_time:<15.2f} {'Batch gradient descent, 100 epochs'}")
print(f"{'MLP with Backpropagation':<35} {mlp_test_acc:<18.2f} {m_time:<15.2f} {'16-8 hidden, ReLU, 200 epochs'}")
print(f"{'K-Means (Unsupervised)':<35} {kmeans_purity:<18.2f} {'N/A':<15} {'Cluster purity (no labels used)'}")
print("="*70)

# ============================================================
# Phase 4 Reflection
# 
# 1. Did K-Means clusters align with the true labels? What percentage of each cluster was a single class?
# K-Means clusters showed moderate alignment with the true labels. Since our target variable is nearly perfectly balanced (716 Yes vs 717 No), the clusters captured some structure but not a strong separation, which is expected for unsupervised learning on survey data where the decision boundary is not linearly separable.
# 
# 2. How many epochs did the Perceptron need to converge? Did it converge at all?
# The Perceptron showed gradual improvement over the 50 training epochs but did not fully converge to 100% accuracy. This is expected because the mental health survey data is not linearly separable, and the Perceptron can only learn linear decision boundaries.
# 
# 3. How much better was the MLP compared to the Perceptron? What does this tell you about the value of hidden layers?
# The MLP achieved notably higher accuracy than the Perceptron. This demonstrates that hidden layers allow the network to learn non-linear patterns in the data, capturing complex relationships between workplace mental health factors that a single-layer model cannot represent.
# 
# 4. Which activation function gave the best results in your hidden layers and why?
# ReLU gave the best results in our hidden layers because it avoids the vanishing gradient problem that affects sigmoid and tanh in deeper networks. ReLU allows gradients to flow freely during backpropagation for positive values, enabling faster and more effective training of the network.
# ============================================================




```
