Step 5: Object-Oriented Representation

```python
# we define our class DataRecord
class DataRecord:
    def __init__(self,record_id,features,label):
        self.record_id=record_id
        self.features=features
        self.label=label
    def Display(self):
        print(f"Record ID: {self.record_id}\n")
        print(f"Features: {self.features}\n")
        print(f"Label: {self.label}\n")
        print("-"*15)
#we make an empty python list to store our objects 
object_list=[]
#for loop to iterate over 5 rows
for row_index in range(5):
    #we use iloc() to grab the row with index row_index which we just defined in our for loop (eg= row0,row1,..etc) inside our row variable
    row=df.iloc[row_index]
    #we store the index of our row as record id in a new variable 
    obj_id=row_index
    #we store the label as our target column since that column is our main focus 
    obj_label=row[target_col]
    #we store rest of the columns as a separate feature in the form of a dictionary to record feature and the response of the object
    #we also drop target_col since we already stored it as our label 
    obj_features=row.drop(target_col).to_dict()
    #we pass the values into the constructor of our object 
    new_record_object=DataRecord(record_id=obj_id,label=obj_label,features=obj_features)
    #we add that item into our list of objects (object_list)
    object_list.append(new_record_object)
#we iterate over our object list and call display for each object
for obj in object_list:
    obj.Display()
```

Step 6: Build a Graph from Your Data

```python
#we need to build a graph between 2 cols we've used are given below in col1 and col2
#the relation ship the above columns represent is how soughting help for mental health affects work if you have 
col1="Have you ever sought treatment for a mental health issue from a mental health professional?"
col2="If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?"
#we define an empty dictionary for our graph the dictionary's key will be the unique values of each column and the values will be 
#a list which will contain all the relationships col1 has with col2 i.e neighbors eg= col1=Yes and col2=No is one relationship 
#col1=Yes and col2=Rarely is another one so node col1 has neighbors No and Rarely 
graph={}
for index in range(len(df)):
  #we grab the current row in our dataset
    row=df.iloc[index]
 #since col1 also has 1 or 0 as vals we convert it into str to check it
    raw_val=str(row[col1]).strip()
    #if 1 we take it as a yes
    if(raw_val=='1')or (raw_val=='1.0'):
        val='yes'
    #else we take it as a no
    else:
        val='no'
    #we append these strings to differ diff types of inference and yes/no to make it sound more meaningfull
    node_A=val.strip().title()+"(Sought Treatment)"
    node_B=str(row[col2]).strip().title()+"(Interference)"
    #now we simply fill our dictionary for initial responses to setup nodes
    if (node_A not in graph):
        graph[node_A]=[]
    if (node_B not in graph):
        graph[node_B]=[]
    #we setup neighbors of each node
    if (node_B not in graph[node_A]):
         graph[node_A].append(node_B)
    if(node_A not in graph[node_B]):
        graph[node_B].append(node_A)
#we print the graph now
print("Dataset Graph\n")
for node,connected_nodes in graph.items():
    print(f"{node} is connected to -->{connected_nodes}")
#we calculate the total number of nodes and edges in the  graph
print(f"the total number of edges in the graph are {len(graph)}\n")
print(f"the total number of edges in the graph are {sum(len(connections) for connections in graph.values())//2}\n")
```
